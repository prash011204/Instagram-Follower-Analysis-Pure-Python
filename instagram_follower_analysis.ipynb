# OPENAI's INSTAGRAM FOLLOWERS ANALYSIS

#### Reading file 

```python
with open("finaldata.txt", encoding='utf-8') as f:
    data = f.read()
```

```python
print(data[:1000])   # First 1000 characters
```

```python
chunks = data.split("\n\n")
```

#### Creating Structural Dataset

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split("\n")
    username = sep_chunk[0]
    no_of_posts = int(sep_chunk[1].split(" post")[0].replace(",", ""))
    no_of_followers = float(sep_chunk[2].split(" followers")[0].replace(",", "").replace("K", "").replace("M", ""))
    if("K" in sep_chunk[2]):
        no_of_followers = int(no_of_followers * 1000)
    elif("M" in sep_chunk[2]):
        no_of_followers = int(no_of_followers * 1000000)
    no_of_followers = int(no_of_followers)
    no_of_following = float(sep_chunk[3].split(" following")[0].replace(",", "").replace("K", "").replace("M", ""))
    if("K" in sep_chunk[3]):
        no_of_following = int(no_of_following * 1000)
    elif("M" in sep_chunk[3]):
        no_of_following = int(no_of_following * 1000000)
    no_of_following = int(no_of_following)
    name = sep_chunk[4]
    type_of_page = sep_chunk[5]
    bio = "\n".join(sep_chunk[6:])

    # print(username, no_of_posts, no_of_followers, no_of_following, name, type_of_page, bio)
    return {"username":username, "no_of_posts":no_of_posts, "no_of_followers":no_of_followers, "no_of_following":no_of_following, "name":name, "type_of_page":type_of_page, "bio":bio}


analysis = []

for chunk in chunks:
    analysis.append((parse_chunk(chunk)))

# for chunk in chunks:
#     print(parse_chunk(chunk))


```

#### Identify Key Accounts Metrics

```python
maximum_post = analysis[0]['no_of_posts']
max_index = 0
for i in range(0, len(analysis)):
    if maximum_post < analysis[i]['no_of_posts']:
        maximum_post = analysis[i]['no_of_posts']
        max_index = i

print(f"MAX POSTS are {maximum_post} of: ", analysis[max_index], "INDEX_NUMBER: ", max_index, "\n\n")
```

```python
maximum_followers = analysis[0]['no_of_followers']
max_index_followers = 0
for i in range(0, len(analysis)):
    if maximum_followers < analysis[i]['no_of_followers']:
        maximum_followers = analysis[i]['no_of_followers']
        max_index_followers = i

print(f"MAX FOLLOWERS are {maximum_followers} of: ", analysis[max_index_followers], "INDEX_NUMBER: ", max_index_followers, "\n\n")
```

```python
maximum_following = analysis[0]['no_of_following']
max_index_following = 0
for i in range(0, len(analysis)):
    if maximum_following < analysis[i]['no_of_following']:
        maximum_following = analysis[i]['no_of_following']
        max_index_following = i

print(f"MAX FOLLOWINGS are {maximum_following} of: ", analysis[max_index_following], "INDEX_NUMBER: ", max_index_following, "\n\n")
```

```python
Categories = []

for i in range(len(analysis)):
    Categories.append(analysis[i]["type_of_page"])

Categories = set(Categories)
Categories = list(Categories)

print(Categories, "There are", len(Categories), "Categories")
```

```python
print("TOP 10 MOST FOLLOWED ACCOUNTS\n")

analysis.sort(key=lambda x: x["no_of_followers"], reverse=True)
for i in range(0, 10):
    print(analysis[i]['username'], "followers = ",  end=" ")
    print(analysis[i]['no_of_followers'])
```

#### Categories distribution analysis

```python
# Count accounts in each category
category_count = {}

for account in analysis:
    category = account["type_of_page"]

    if category in category_count:
        category_count[category] += 1
    else:
        category_count[category] = 1

# Sort by number of accounts (highest to lowest)
sorted_categories = sorted(category_count.items(), key=lambda x: x[1], reverse=True)

# Display
print("Category-wise Account Count")
print("-" * 35)

for category, count in sorted_categories:
    print(f"{category}: {count}")
```
